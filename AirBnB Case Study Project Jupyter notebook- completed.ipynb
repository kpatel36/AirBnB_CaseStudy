```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb
```

```python
air_bnb.columns

air_bnb2=air_bnb[:]
```

```python
air_bnb2.groupby(['host_id','host_name']).count()[['calculated_host_listings_count']].sort_values('host_id').head()
```

```python
# Which hosts are the busiest and why?

# if you go based on number of reviews for each host: Maya, Brooklyn*Breakfast-Len-, Danielle, Yasu & Akiko, and Brady are the busiest hosts since they have the most reviews 
air_bnb2.groupby(['host_id', 'host_name']).sum()[['number_of_reviews']].sort_values('number_of_reviews',ascending=False).head()

# if you go based on number of reviews per month for each host, Sonder(NYC), Corporate Housing, Jeremy & Laura, Kazuya, Vida are the busiest hosts 
air_bnb2.groupby(['host_id','host_name']).count()[['reviews_per_month']].sort_values('reviews_per_month',ascending=False).head()
```

```python
# How many neighborhood groups are available and which shows up the most?
air_bnb2.groupby('neighbourhood_group').count().sort_values('neighbourhood', ascending=False)

# There are five total neighbourhood groups and Manhattan shows up the most
```

```python
# Are private rooms the most popular in manhattan?
# air_bnb2.groupby(['neighbourhood_group', 'room_type']).count()[['neighbourhood']]
# air_bnb2
air_bnb2[air_bnb2['neighbourhood_group']=='Manhattan'].groupby('room_type').count()[['id']]

# No, "Entire Home/Apt" rentals are more popular than "Private Rooms" in Manhattan
```

```python
# Which hosts are the busiest based on their reviews?
air_bnb2.columns
# air_bnb2.groupby(['host_name','number_of_reviews']).count()[['reviews_per_month']].sort_values('number_of_reviews', ascending=False).head()
air_bnb2.groupby(['host_name','host_id']).sum()[['number_of_reviews']].sort_values('number_of_reviews', ascending=False).head()
```

```python
#Which neighorhood group has the highest average price?
air_bnb2.groupby('neighbourhood_group').mean()[['price']].sort_values('price', ascending=False).head(1)
```

```python
# Which neighborhood group has the highest total price?
air_bnb2.groupby('neighbourhood_group').sum()[['price']].sort_values('price',ascending=False).head(1)

```

```python
#Which top 5 hosts have the highest total price?
air_bnb2.groupby(['host_name', 'neighbourhood_group']).sum()[['price']].sort_values('price', ascending=False).head(1)
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
air_bnb2[(air_bnb2['availability_365']==0) & (air_bnb2['number_of_reviews']>=100)].groupby(['host_id','host_name']).sum()[['number_of_reviews']].sort_values('number_of_reviews')

# There are currently 147 hosts with 0 availablity and more than 100 reviews
```

```python
# What host has the highest total of prices and where are they located?

# if you go based on price per rental
air_bnb2.groupby(['host_id','host_name','neighbourhood'])[['price']].sum().sort_values('price',ascending=False).head()
```

```python
# When did Danielle from Queens last receive a review?
air_bnb2[['host_name','neighbourhood_group','neighbourhood','last_review']][(air_bnb2['host_name']=="Danielle") & (air_bnb2['neighbourhood_group']=='Queens')].sort_values('last_review',ascending=False)
```
